## Introduccion a la Regresion Lineal.

```python
from sklearn.linear_model import LinearRegression
```

# Que es una regresion lineal?
Una regresin intenta predecir un valor numrico a partir de otros datos.

Ejemplos:

- Predecir ingresos segn edad.
- Predecir precio segn tamao.

Aqu usaremos regresin lineal: el tipo ms simple y comn.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x = df_clean["edad"], y = df_clean["ingresos"])
plt.xlabel("Edad")
plt.ylabel("Ingresos")
plt.title("Relacion entre Edad y Ingresos")
plt.show()
```

```python
#MATRIZ DE CORRELACION
#df_clean.corr()
df_clean[["edad", "ingresos"]].corr()

#0.0 -> nada
#0.3 -> dbil
#0.50.7 -> moderada
#0.8+ -> fuerte
```

```python
X = df_clean[["edad"]]
y = df_clean["ingresos"]
modelo = LinearRegression()
modelo.fit(X, y)
```

Tiene relacion con la formula de la pendiente. Qu hace fit? Busca los valores de m (pendiente) y b (intercepto) que mejor ajustan tus datos. Minimiza el error entre los ingresos reales y los ingresos predichos. Despus de fit, el modelo ya aprendi la relacin: ingresosbedad + a formula: mx + b

```python
edad_nueva = [[30]]
prediccion = modelo.predict(edad_nueva)
prediccion
```

```python
# Podemos evaluar el modelo
from sklearn.metrics import r2_score, mean_squared_error
y_pred = modelo.predict(X) #Son los ingresos que el modelo cree segun la edad
#ingresos = 0 + 1  edad
#0  intercepto
#1  coeficiente de edad

r2 = r2_score(y, y_pred)
#0.3 - 0.5 aceptable, pero siempre es mejor tener un modelo 0.6 - 0.8
# o muy bueno > 0.8
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
print(r2)
# Regresion logistica se suele usar -> Accuray, Precision, F1 Score, Recall,
```
