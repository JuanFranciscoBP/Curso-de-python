## Operaciones Matematicas

```python
# 4to nivel de importancia
# Suma ordinaria
print("Suma ordinaria: ")
print(7 + 4) #un 11
variable1,variable2 = 40,90 #un 130
print(variable1 + variable2)

#Resta ordinaria
print("Resta ordinaria: ")
print(4 - 5) # -1

#Declaracion de variables
v1 = 13.30
v2 = 12.10
resultado = v2 - v1 # 1.2
print(resultado)
```

```python
# 3er nivel de importancia
#Multiplicacion ordinaria
print("Multipliacion ordinaria")
M1, M2 = 4,3.14
print(M1 * M2) #6.28
#Division ordinaria
print("-------------")
print("Division ordinaria")
M1,M2,M3,M4,M5,M6 = 3,6,1,3,4,5
print(M1 / M2)
print(M3 / M4)
print(M5 / M6)
#Division de Piso
print("-------------")
print("Division de Piso")
print(M1 // M2)
print(M3 // M4)
print(M5 // M6)
print( 6 / 2 *3)
```

```python
#Potencias
print("Potencias")
print(2 ** 4) #16
P1,P2 = 7,2
print(P1 ** P2) #49
```

```python
# Area de un circulo
pi = 3.14
r = 4
A = pi * (r ** 2)
A
```
