JSON es un formato para guardar y transmitir datos. Es muy en APIs, archivos de configuracion y almacenamiento de informacion.
(se parece a un diccionario de Pyhton, pero es solo texto).

Para que sirve?
  - Sirve para convertir objetos de python -> texto JSON y tambien texto JSON -> objetos de Python

  Esto se llama:
    - Serializar: Convertir un objeto de Python en texto JSON
    - Deserializar: Convertir un texto JSON a un objeto usable de Python

```python
# Funciones principales:
#1) json.dums(obj): Convierte un diccionario u objeto de Pyhton en un string JSON
import json
obj = {"Nombre": "Ana"
, "edad": 25}
print(type(obj))
s = json.dumps(obj, ensure_ascii= False)
print(s)
print(type(s))
#
```

```python
# json.loads(texto_json): Convierte un string JSON en un diccionario Python
restaurar = json.loads(s)
print(restaurar)
print(type(restaurar))
# Al venir de una pagina/persona/archivo desconocido

'''
Pueden contener:

Datos malinformados

Informacion inesperada

Campos faltantes

Datos demasiado grandes

va incorrectas que rompan tu programa

'''
```

```python
'''#UTF-8 es la abreviatura de "8-bit Unicode Transformation Format" y se trata de un formato de codificación de caracteres Unicode e ISO 10646 que utiliza símbolos de longitud variable, permitiendo representar cualquier carácter Unicode con una eficiencia que lo hace ideal para internet.'''
with open("datos.json", "w", encoding="utf-8") as f:
   json.dump(obj, f, ensure_ascii=False, indent=4)
```

## Numpy
Provee el tipo ndarray , operaciones vectorizadas, brodcasting, utilizades de algenra lineal y funciones matematicas en arrays. Es la base sobre la que se apoyan muchas librerias cientificas en Pyhton.

### Nota IMPORTANTE: El arreglo de Numpy no es igual a la lista de Python, este NO puede tener diferentes tipos de datos en el mismo

```python
#pip install numpy
import numpy as np
A = np.array([[1,2,3],[4,5,6]]) #array (2,3)
v = np.array([10,20,30]) #array (3)
print(A + v)
print("-----------")
prueba = [1,2,3]
prueba2 = [1,2,3]
print(prueba + prueba2)
```

```python
# Mas funciones de Numpy
z = np.zeros((3,3)) #array (3,3)
print(z)
o = np.ones((2,4))
print(o)
r = np.random.rand(3,3)
print(r)
```

```python
# Rangos numericos
nums = np.arange(0,10,2) # saca los valores 0 y 10 de 2 en 2
print(nums)
x = np.linspace(0,1,5) # saca 5 valores espaciados de manera "equitativa" entre 0 y 1
print(x)
```

```python
a = np.array([1,2,3,4,5])
b = np.array([6,7,8,9,10])
print(a + b) #suma
print(a - b) #resta
print(a * b) #multiplicacion
print(a / b) #division
print(a ** b) #potencia
```

```python
print(a.dot(b)) #producto entre dos matrices
c = np.array([[1,2,3],[4,5,6]])
d = np.array([[10,20],[30,40],[50,60]])
#1 *10 + 2 * 30 + 3 * 50
#
#
#
print(c.dot(d))
```

```python
data = np.array([1,2,3,4,5])
print(data.max()) #minimo
print(data.min()) #maximo
print(data.sum()) #suma
print(data.mean()) #promedio
print(data.std()) #desviacion estandar
```
