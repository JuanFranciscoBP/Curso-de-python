# Estructuras de Datos
Las estructuras de datos son formas de organizar informacion para trabajar con ellas de manera eficiente.

- Listas -> ordenadas, modificadas, permiten duplicados
- Tuplas -> ordenadas, inmutables, permiten duplicados
- Sets -> no ordenados, sin duplicados
- Diccionarios -> pares clave: valor

Nota:

- Pilas (FILO -> First in, last out)
- Colas (FILO -> Firsr in, First out)
- Listas enlazadas (o doblemente enlazadas)

## Listas
Las listas son colecciones ordenadas que podemos modificar y almacenan elementos de distintos tipos de datos.

```python
lista_de_compras = ["manzanas","pan","leche","huevos","mantequilla","queso","rabano"]
print(lista_de_compras)
```

```python
print(lista_de_compras[6])
```

```python
# Agregar un elemento a la lista .append()
lista_de_compras.append("sopa")
lista_de_compras
```

```python
# Agregar un elemento en una posicion especifica.
lista_de_compras.insert(1,"platano")
lista_de_compras
```

```python
print(lista_de_compras)
```

```python
# Para eliminar un elemento de una lista existe el remove y el pop
lista_de_compras.remove("platano")
print(lista_de_compras)
```

```python
lista_de_compras.pop() #Sin nada dentro del parentesis borra el ultimo elemento
lista_de_compras.pop(7) # o puede agarrrar la posicion que quieras (pero iniciando desde 0)
# Nota: te retorna el numero eliminado
```

```python
# Podemos recorrer el arreglo por medio de un for
for elementos in lista_de_compras:
    print(elementos)
```

```python
# len() saca el tamaño de la lista
for i in range(len(lista_de_compras)):
    print(f"{i + 1 } - {lista_de_compras[i]}")
```

```python
promedios = [55.6,97.5,23.6,97.5,72.4,100]
print(promedios)
promedios.sort() #sort() sirve para ordenar la lista de manera ascendente
print(promedios)

# ordenar una lista de manera descendente
promedios.sort(reverse=True)
print(promedios)
```

```python
#Podemos ingresar al ultimo elemento de una lista
print(promedios[-1])
```

```python
# Encontrar el mayor y el menos de una lista
print(max(promedios))
print(min(promedios))
```

```python
#Declaracion y uso de matrices
calificaciones = [[100,88,93],[99,56,72]]
i = 1
for cali_i in calificaciones:
  print(f"\nAlumno {i}")
  j = 1
  for cali_j in cali_i:
    print(f"cali {j}: {cali_j}")
    j += 1
  j = 0
  i += 1
```

## SETS
Los sets son colecciones SIN orden y que NO permiten elementos duplicados.

```python
nombre = {"Angel", "Emiliano", "Diana", "Juan", "Juan"}
print(nombre)
print(type(nombre))
```

```python
# Add
nombre.add("Leonardo")
nombre
```

```python
print(nombre[2])
```

```python
nombre[2] = "Axel"
```

```python
# Union
matutino = {20322,22183,12312,34324}
vespertino = {23243,343243,43242434,34424}
# Union pero con .union
total_matriculados = matutino.union(vespertino)
print(total_matriculados)
# Union pero con |
total_matriculados = matutino | vespertino
print(total_matriculados)
```

```python
# Interseccion
ca = {1234, 123,1232,123123,312312,3123,32324}
al = {1234,23,323,434,545}
#interseccion con .interseccion()
ca_al = ca.intersection(al)
print(ca_al)
# Interseccion con &
ca_al = ca & al
print(ca_al)
```

```python
# Diferencia
# Diferencia usando .difference
solo_calculo = ca.difference(al)
print(solo_calculo)
# Diferencia usando -
solo_calculo = ca - al
print(solo_calculo)
```

```python
nombre = {"Angel", "Emiliano", "Diana", "Juan", "Juan"}
#remove, pop
nombre.remove("Angel") #Puedes eliminar el que tu quieras
print(nombre)
nombre.pop() # Elimina un elemento de manera aleatoria
print(nombre)
```

```python
# Eliminar todo el SET
nombre.clear()
nombre
```

## Diccionarios
Los diccionarios son colecciones no ordenadas de pares clave: valor

```python
alumno = {
    #Clave    Valor
    "nombre": "Emiliano Cepeda",
    "edad" : 23,
    "carrera": "Ingenieria en Software",
    "semestre": 9,
    "promedio": 88,
    "calificaciones" : [100,94,94,97,100]
}
print(type(alumno))
alumno
```

```python
print("El alumno", alumno["nombre"],"esta estudiando", alumno["carrera"])
```

```python
# Agregar al diccionario
alumno["area preferida"] = "Sistemas"
alumno
```

```python
for valor in alumno:
  print(alumno[valor])
```

```python
for aspecto, valor in alumno.items():
  #devuelve una vista de todos los pares clave : valor del diccionario
  print(aspecto, ":", valor)
```

```python
# El metodo .clear() tambien se puede usar aqui.
# Tenemos el .get que nos ayuda a obtener el valor (value) de cierta clave (key)
print(alumno.get("nombre"))
```

```python
alumno_1 = alumno.items()
print(alumno_1)
print(type(alumno_1))
alumno_lista = list(alumno_1)
print(alumno_lista)
print(type(alumno_lista))
```

```python
#El metodo .values() sirve para obtener los valores del diccionario, tambien se puede usar en la iteraccion
alumno_V = alumno.values()
print(type(alumno_V))
alumno_V
```

```python
if "carrera" in alumno:
  print("Existe carrera")
else:
    print("No existe carrera")

if("dinero") in alumno:
  print("Alumno tiene dinero")
else:
  print("Alumno no tiene dinero")
```

```python
#.copy() puedes copiar todo el diccionario en una nueva variable
# Update -> Sirve para cambiar los (valores) (.update())
alumno.update({"semestre":6})
alumno
```

```python
# Lista de productos
productos = ["manzana", "banana", "leche", "pan"]  # <-- LISTA

# Tupla de precios
precios = (10, 5, 20, 15)  # <-- TUPLA

# Diccionario de inventario: producto -> {"precio": x, "stock": y}
inventario = {
    productos[i]: {"precio": precios[i], "stock": 10}
    for i in range(len(productos))
}
# <-- DICCIONARIO
# Clave: nombre del producto
# Valor: otro diccionario con "precio" y "stock"

# Set de productos en oferta
ofertas = {"manzana", "pan"}  # <-- SET

# Mostrar productos
for p in productos:
    msg = "OFERTA!!!" if p in ofertas else "Grandes precios!"
    info = inventario[p]
    print(f"{p.title()} - ${info['precio']} - Stock: {info['stock']} -> {msg}")

# Compra simple
compra = input("\n¿Qué producto quieres comprar? ").strip().lower()

if compra in inventario and inventario[compra]["stock"] > 0:
    inventario[compra]["stock"] -= 1
    print(f"Compraste 1 {compra}. Stock restante: {inventario[compra]['stock']}")
else:
    print("Producto no disponible.")
```
