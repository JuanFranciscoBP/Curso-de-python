- Puede acceder y modificar los atributos de ese objeto
- Recibe como primer parametro a self

```python
# (ya lo hemos usado)
class tuClase:
    def metodo(self):
        pass
#metodo -> método de instancia
#self -> referencia al objeto que llama al método
```

```python
# Calculadora:
class Calculadora:
    def __init__(self,num1,num2):
        self.num1 = num1
        self.num2 = num2
    def sumar(self):
        print(self.num1 + self.num2)
    def resta(self):
        print(self.num1 - self.num2)
    def multiplicar(self):
        print(self.num1 * self.num2)
    def dividir(self):
        if self.num2 == 0:
            print("Numero no divisible")
        else:
            print(self.num1/self.num2)
Calculadora = Calculadora(7,0)
Calculadora.dividir()
Calculadora.sumar()
Calculadora.multiplicar()
```

```python
# Getters y Setters
class Cuenta:
    def __init__(self,saldo):
        self.__saldo = saldo
    @property
    def saldo(self): #Getter
        #Permite leer el valor de __saldo usando c.saldo
        return self.__saldo
    @saldo.setter #Setter 
    def saldo(self, valor): 
        # se ejecuta cuando hacemos c.saldo = valor
        self.__saldo = valor
c = Cuenta(300)
print(c.saldo)
c.saldo = 4000
```

## 7.- Herencia y Abstraccion

```python
# Herencia
class Vehiculo:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        
    def descripcion(self):
        return f"{self.marca}:{self.modelo}"
    def mover(self):
        return "El vehiculo se esta moviendo..."
#------------------------------------------------------------------------
class Coche(Vehiculo): #subclase 1 (clase hija)
    def mover(self):
        return "El coche esta conduciendo"
#------------------------------------------------------------------------
class Moto(Vehiculo):
    def mover(self):
        return "La moto avanza sobre dos ruedas"

coche = Coche("Toyota", "Corolla")
coche2 = Vehiculo("BYD", "Dolphin")
moto = Moto("Yamaha", "R6")
print(coche.mover())
print(coche2.mover())
print(moto.mover())
#aun que no volvimos a declarar el metodo dentro de descripcion, podemos seguir usandolo ya que lo heredo del padreprint(coche.mover())print(moto.descripcion())print(moto.mover())
```

Porque estas componiendo (armando) un objeto usando otros objetos. Es un principio de POO donde:

## 9.- Composicion vs herencia
Esto se va a responder con pura:
No significa que alguna sea mejor que otra simplemente que aveces una te puede convenir mas que otra:

Herencia: un gato ES un animal (no "Un gato TIENE un animal")

Composicion: Un a persona TIENE una cuenta (no "Una persona Es un Banco")

```python
'''Ejercicio Aplicando todo:Queremos modelar un pequeño sistema para vehículos.
Habrá distintos tipos:CocheMotoBicicletaTodos son “vehículos” (clase vehiculo), 
por lo tanto comparten características básicas, pero cada uno se mueve distinto, 
debes poner una clase llamada motor la cual use Composición, opcional: Usa Polimorfismo'''
# Clase base
class Vehiculo:
    def __init__(self, marca):
        self.marca = marca

    def mover(self):
        raise NotImplementedError("Este método DEBE ser implementado.")


# Clase Motor (COMPOSICIÓN)
class Motor:
    def __init__(self, caballos):
        self.caballos = caballos

    def potencia(self):
        return f"{self.caballos} HP"


# Clase Coche
class Coche(Vehiculo):
    def __init__(self, marca, caballos):
        super().__init__(marca)
        self.motor = Motor(caballos)  # COMPOSICIÓN

    def mover(self):  # POLIMORFISMO
        return f"El coche {self.marca} avanza con motor de {self.motor.potencia()}"


# Clase Moto
class Moto(Vehiculo):
    def __init__(self, marca, caballos):
        super().__init__(marca)
        self.motor = Motor(caballos)  # COMPOSICIÓN

    def mover(self):  # POLIMORFISMO
        return f"La moto {self.marca} acelera con motor de {self.motor.potencia()}"


# Clase Bicicleta
class Bicicleta(Vehiculo):
    def mover(self):  # POLIMORFISMO
        return f"La bicicleta {self.marca} avanza con pedaleo humano"


# Función que prueba el polimorfismo
def probar_vehiculos(lista):
    for v in lista:
        print(v.mover())


# Lista de vehículos (sí, se pueden hacer listas de clases)
vehiculos = [
    Coche("Toyota", 150),
    Moto("Yamaha", 90),
    Bicicleta("Trek")
]

probar_vehiculos(vehiculos)
```
